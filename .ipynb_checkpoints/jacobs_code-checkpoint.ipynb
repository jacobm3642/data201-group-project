## API Function

This function allows you to call the WHO API to retrieve the raw JSON data.

    - Arg indicator: Takes the indicator defined on the WHO website to specify which data to retrieve.
    - Arg use_athena: Takes 1 or 0. Use 1 to indicate that it will be using the Athena API, and 0 for the Azure API.


```r
get_health_data <- function(indicator, use_athena) {
    # alows both api to be used
    if (use_athena == 1) {
        base_url <- "http://apps.who.int/gho/athena/api/"
        url <- paste0(base_url, "GHO/", indicator, ".json")
    } else if (use_athena == 0) {
        base_url <- "https://ghoapi.azureedge.net/api/"
        url <- paste0(base_url, indicator)
    } else {
        cat("Invalid argument use 1 for Athena API or 0 for Azure API.")
        return(NULL)
    }
    # tell you where you are geting your data
    cat("URL:", url, "\n") 
    
    response <- GET(url)

    # fails non fataly if the api errors and tells you why 
    if (http_error(response)) {
        cat("L1 Error:", http_status(response)$reason, "\n")
        return(NULL)
    } else {
        return(content(response, "parsed"))
    }
}
```

## API cleaning Function

This function allows you to get a dataframe from the Azure API in a semi-clean usable format.

    - Arg indicator: Takes the indicator defined on the WHO website to specify which data to retrieve.
    - Dependence: It requires the get_health_data to be loaded as it builds on top of it.

# Note:
the API is down as of the 25th of october 2023 but it will come back so im keeping this in 

```r
get_ind_data <- function(indicator) {
    # this defineds two list so that the index of the type name is the index of what the col shoud be coverted into
    col_names_dict <- c("COUNTRY", "YEAR", "SEX", "AGEGROUP", "ENVCAUSE")
    col_names_type <- c("factor", "int", "factor", "factor", "factor")
    
    # gets the raw json from the api
    ind <- get_health_data(indicator, 0)
    ind <- ind %>% gather_array %>% spread_all # json to dataframe convertion
    col_names <- names(ind)
    # loops thought the cols
    for (col in col_names){
        # checks if the col name contaens the substring "Type"
        if (grepl("Type", col)) {
            type_col <- col
            # the col the type col refrences is the name type col without the substring "Type"
            ref_col <- gsub("Type", "", col)
            type <- ind[[type_col]][1]
            indices <- which(col_names_dict == type)
            # if the type col contanes NA it is unused so its removed
            # else if length(indices) > 0 means there are some value for indices
            # and error check if the indices are NA (it happened once so theres a check for it)
            if (is.na(type)) {
                ind <- ind %>% select(-c(type_col, ref_col))
            } else if (length(indices) > 0 && !is.na(col_names_type[indices[1]])) {
                # tell the sys what type the col needs to be converted into
                type_label <- col_names_type[indices[1]]
                # why doesn't r have switch statements i hate this style but it works
                if (type_label == "factor") {
                    ind <- ind %>% mutate(!!ref_col := as.factor(!!sym(ref_col)))
                    ind <- ind %>% rename(!!type := !!sym(ref_col))
                    ind <- ind %>% select(-type_col)
                } else if (type_label == "int") {
                    ind <- ind %>% mutate(!!ref_col := as.integer(!!sym(ref_col)))
                    ind <- ind %>% rename(!!type := !!sym(ref_col))
                    ind <- ind %>% select(-type_col)
                } 
            }
        }
    }
    # brings api data style inline with the rest of the project  
    ind <- ind %>% rename(country_codes = COUNTRY)
    ind <- remove_all_na_cols(ind)
    return(ind)
}

# i stole this from one of my old projects
remove_all_na_cols <- function(data) {
    na_counts <- colSums(is.na(data))
    na_cols <- names(na_counts[na_counts == nrow(data)])
    data <- data[, !names(data) %in% na_cols]
    return(data)
}
```

## local cleaning Function

This function allows you clean a local csv from the WHO in a simmler wat as the API cleaning Function.

    - Arg data: Takes a data frame idealy from read.csv.

# Note:
This is pritty much the same as the API cleaning Function but with some things changed as the style of the downloaded csv are not the same as the api

```r
get_ind_data_local <- function(data) {
    # mostly it just doesnt do the api stuff
    
    # this is one of the changes the download uses spaces and isnt all caps
    col_names_dict <- c("Country", "Year", "Sex", "Age Group")
    col_names_type <- c("factor", "int", "factor", "factor")
    ind <- data
    col_names <- names(ind)
    # the loop is all the same as above
    for (col in col_names){
        if (grepl("Type", col)) {
            type_col <- col
            ref_col <- gsub("Type", "", col)
            type <- ind[[type_col]][1]
            indices <- which(col_names_dict == type)
            if (is.na(type)) {
                ind <- ind %>% select(-c(type_col, ref_col))
            } else if (length(indices) > 0 && !is.na(col_names_type[indices[1]])) {
                type_label <- col_names_type[indices[1]]
                # why doesn't r have switch statements i hate this style but it works
                if (type_label == "factor") {
                    ind <- ind %>% mutate(!!ref_col := as.factor(!!sym(ref_col)))
                    ind <- ind %>% rename(!!type := !!sym(ref_col))
                    ind <- ind %>% select(-type_col)
                } else if (type_label == "int") {
                    ind <- ind %>% mutate(!!ref_col := as.integer(!!sym(ref_col)))
                    ind <- ind %>% rename(!!type := !!sym(ref_col))
                    ind <- ind %>% select(-type_col)
                } 
            }
        }
    }
    # the name of the Country code col id diffrent
    ind <- ind %>% rename(Country_code = SpatialDimValueCode)
    ind <- remove_all_na_cols(ind)
    return(ind)
}
```

## Boxplot for BMI versus whether a country had a plan to combat diabetes from 2013 to 2015.

```r
# load the data (bmi is used in all the graphs but is just loaded here)
bmi <- get_ind_data_local(read.csv("./bmi_data.csv"))
diabetes_plan <- get_ind_data_local(read.csv("./diabetes_data.csv"))

# its a yes or no so factor makes cents
diabetes_plan <- diabetes_plan %>% mutate(Value = as.factor(Value))

diabetes_bmi <- merge(diabetes_plan, bmi, by = c("Country_code", "Year"))

# select the data we will use
diabetes_bmi <- diabetes_bmi %>% select(Value.x, FactValueNumeric, Country_code, Year)

# give it good names
diabetes_bmi <- diabetes_bmi %>% rename(plan_res =Value.x,bmi = FactValueNumeric)

diabetes_bmi <- diabetes_bmi %>% filter(plan_res == "Yes" | plan_res == "No")

diabetes_bmi <- na.omit(diabetes_bmi)

diabetes_bmi_2013 <- subset(diabetes_bmi, Year == 2013)
diabetes_bmi_2015 <- subset(diabetes_bmi, Year == 2015)

plot_2013 <- ggplot(diabetes_bmi_2013, aes(x = factor(plan_res), y = bmi)) +
    geom_boxplot() +
    labs(title = "Boxplot for BMI in 2013",
         x = "If the country had a plan to combat diabetes",
         y = "BMI by country") +
    theme_minimal()

plot_2015 <- ggplot(diabetes_bmi_2015, aes(x = factor(plan_res), y = bmi)) +
    geom_boxplot() +
    labs(title = "Boxplot for BMI in 2015",
         x = "If the country had a plan to combat diabetes",
         y = "BMI by country") +
    theme_minimal()

# combine the plots
grid.arrange(plot_2013, plot_2015, ncol = 2)
```

I found it interesting the significant change in the interquartile range, countries that did not have a plan to combat diabetes in 2015 vs 2013. Because while both of them went up, the mean BMI went up across the world, they saw a significant change in the interquartile range, which I just thought was a pretty interesting result although this may be a result of a smaller number of countries not having a plan to combat diabetes. So it skews the data a bit.

## The availability of diabetes testing versus the mean BMI by country boxplot.

```r
# load new data
availability_of_diabetes_testing <- get_ind_data_local(read.csv("./H1MseuNv.csv"))

availability_of_diabetes_testing <- availability_of_diabetes_testing %>% mutate(Value = as.factor(Value))

bmi_avail <- merge(availability_of_diabetes_testing, bmi, by = c("Country_code", "Year"))

bmi_avail <- bmi_avail%>% select(Value.x, FactValueNumeric, Country_code, Year)

bmi_avail <- bmi_avail %>% rename(availability_of_diabetes_testing = Value.x,bmi = FactValueNumeric)

# we only want yes and no not no reponce
bmi_avail <- bmi_avail %>% filter(availability_of_diabetes_testing == "Yes" | availability_of_diabetes_testing == "No")

bmi_avail <- na.omit(bmi_avail)

ggplot(bmi_avail, aes(x = availability_of_diabetes_testing, y = bmi)) +
    geom_boxplot() +
    labs(title = "BMI by the availability of diabetes testing",
         x = "The availability of diabetes testing",
         y = "BMI by country") +
    theme_minimal()
```

This intuitively makes sense, which it's nice to know that our intuition is true, that the interquartile range of countries with no available diabetes testing is within the healthy range of BMI. Which I would bet is because Countries that don't have available diabetes testings are likely to have a low GDP. And it was shown by Oscar that countries with a lower GDP tend to have a lower BMI.

## Looking at mean BMI and mean BMI's distance from a healthy midpoint versus the percentage of GDP spent on healthcare.

```r
# this is the middle of the "healthly" bmi range
bmi_helthy_midpoint = 21.7

# makes a col that contanes the distence from the middle of the "healthly" bmi range
bmi <- bmi %>% mutate(dis_from_midpont = abs(FactValueNumeric - bmi_helthy_midpoint))
```

```r
# load new data
health_spend_by_percent_of_gdp <- get_ind_data_local(read.csv("./helth_by_gdb.csv"))

bmi_spend <- merge(health_spend_by_percent_of_gdp, bmi, by = c("Country_code", "Year"))

bmi_spend <- bmi_spend %>% select(FactValueNumeric.y, FactValueNumeric.x, Country_code, Year, dis_from_midpont)

bmi_spend <- bmi_spend %>% rename(bmi = FactValueNumeric.y,spend = FactValueNumeric.x)

bmi_spend <- bmi_spend %>% filter(Year == max(Year))

bmi_spend <- na.omit(bmi_spend)

p1_bmi_gdp <- ggplot(bmi_spend, aes(x = spend, y = bmi)) +
    geom_point()+
    labs(title = "BMI by the percent of gdp spent\non healthcare",
         x = "The percent of gdp spent on health care",
         y = "BMI by country") +
    geom_smooth(method=lm) +
    theme_minimal()

bmi_spend_lt_10 <- bmi_spend %>% filter(spend < 10, !is.null(spend))

p2_bmi_gdp_lt_10 <- ggplot(bmi_spend_lt_10, aes(x = spend, y = bmi)) +
    geom_point()+
    labs(title = "BMI by the percent of gdp spent\non health care cut off at 10% spend",
         x = "The percent of gdp spent on health care ",
         y = "BMI by country") +
    geom_smooth(method=lm) +
    theme_minimal()

p3_bmi_dis_gdp <- ggplot(bmi_spend, aes(x = spend, y = dis_from_midpont)) +
    geom_point()+
    labs(title = "BMIs distence from heathy midpoint by\nthe percent of gdp spent on healthcare",
         x = "The percent of gdp spent on health care",
         y = "BMIs distence from heathy midpoint") +
    geom_smooth(method=lm) +
    theme_minimal()

bmi_spend_lt_10 <- bmi_spend %>% filter(spend < 10)

p4_bmi_dis_gdp_lt_10 <- ggplot(bmi_spend_lt_10, aes(x = spend, y = dis_from_midpont)) +
    geom_point()+
    labs(title = "BMIs distence from heathy midpoint\nby the percent of gdp spent on\nhealthcare cut off at 10% spend",
         x = "The percent of gdp spent on health care ",
         y = "BMIs distence from heathy midpoint") +
    geom_smooth(method=lm) +
    theme_minimal()

# puts the graph togerther in a grid so its quilck to combare them
grid.arrange(p1_bmi_gdp, p2_bmi_gdp_lt_10, 
             p3_bmi_dis_gdp, p4_bmi_dis_gdp_lt_10, 
             ncol = 2)
```

This is partially intuitively true, but I think it's quite interesting that as the percentage of GDP spent on healthcare increases, the mean BMI of that country also increases. I made two graphs for both BMI and distance to BMI because I wanted to see if the extreme values were skewing the linear relationship to be more positive than it actually was. And I did the distance from the healthy midpoint because being underweight is almost as dangerous as being overweight for some healthcare outcomes. And I wanted to sort of see if that changed the relationship.

## overall
I have found that generally, the higher your mean BMI, the more, as a percentage of GDP, you will have to spend on healthcare. As well as having a plan to combat diabetes, while it doesn't necessarily have a significant negative effect on the general trend of increasing BMI, it seems to slow the increase of BMI. And because having a higher mean BMI generally correlates with higher costs, it seems that having a plan to combat diabetes is effective in slowing the raising of mean BMI, but not necessarily in reducing the mean BMI. I would like to note, though, combating diabetes does involve promoting a healthy lifestyle, so it might not have a huge effect on BMI because BMI does not distinguish between fat and muscle.
